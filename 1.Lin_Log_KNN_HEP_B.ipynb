```python
#dataset devided into input and output
X=df.drop('Syndrome ',axis=1)
y=df['Syndrome ']
```

```python
#train test split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
```

```python
y_train=l.fit_transform(y_train)
```

```python
y_test=l.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
```

#using linear regresssion 

```python
lin=LinearRegression()
```

```python
lin.fit(X_train,y_train)
```

```python
prediction=lin.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,prediction)
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,prediction))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction.round()))
```

```python
print(confusion_matrix(y_test,prediction.round()))
```

#ELBOW method

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    error_rate.append(np.mean(pred!=y_test))
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,40),error_rate, color='red',linestyle='--', markerfacecolor='blue', 
        marker='o', markersize=10)
plt.xlabel('Range')
plt.ylabel('Error Rate')
plt.title('Range vs Error Rate')
plt.show()
```
